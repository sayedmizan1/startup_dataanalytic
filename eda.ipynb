```python
scale=[]
for x in dataset["No. of Investors"]:
    if x>0 and x<=10:
        scale.append("Small Scale")
    elif x>10 and x<=25:
        scale.append("Mid Scale")
    else:
        scale.append("Large Scale")

scale
```

```python
dataset.insert(3,"Company Scale",scale)
```

```python
dataset["Company Scale"].value_counts()
```

```python
city_counts=dataset['City'].value_counts()
plt.xticks(rotation=90)
plt.bar(city_counts.index, city_counts.values)
plt.xlabel('City')
plt.ylabel('Number of Startups')
```

```python
top_industries=dataset['Industries'].value_counts().nlargest(10)
plt.bar(top_industries.index,top_industries.values)
plt.xticks(rotation=90)
plt.xlabel('Industry')
plt.ylabel('Number of Startups')
plt.title('Top 10 industries by number of startups')
```

```python
top_ind=dataset.groupby('Industries')['Funding Amount in $'].sum().nlargest(10)
plt.bar(top_ind.index,top_ind.values)
plt.xticks(rotation=90)
plt.xlabel('Industry')
plt.ylabel('Funding Amount($)')
plt.title('Top 10 industries in terms of funding')
```

```python
sby=dataset.groupby('Starting Year')['Funding Amount in $'].sum()
plt.plot(sby.index, sby.values)
plt.xlabel('Year')
plt.ylabel('Total Funding Amount in $')
plt.title('Startup Investments in India')
plt.show()
```

```python
sby=dataset.groupby('City')['Funding Round'].sum().nlargest(10)
plt.xticks(rotation=90)
plt.plot(sby.index, sby.values)
plt.xlabel('City')
plt.ylabel('Total Funding Rounds')
plt.title('No. of Funding rounds according to City')
plt.show()
```
